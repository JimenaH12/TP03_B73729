5. <h4>Matriz H: El Hamiltoniano H se observa como la suma de dos partes:
-  Término cinético, que se aproxima con diferencias finitas (tridiagonal)
-  Término potencial, una matriz diagonal con los valores X_n^2 / 2

Se encuentra la matriz H, tal que H @ psi = E * psi, un problema de autovalores
</h4>

```python
# términos cinéticos
H1 = np.eye(N, k=-1)
H2 = -2 * np.eye(N)
H3 = np.eye(N, k=1)

# segunda derivada multiplicada por factor cinético
H_kinetic = -(H1 + H2 + H3) / (2 * delta_x**2)

# término potencial
H_potential = np.diag(0.5 * X**2)

# Hamiltoniano 
H = H_kinetic + H_potential
```

```python
# verificación (sección de H)

# primeras 10 filas y columnas de H
print("Sección superior izquierda de la matriz H:")
print(H[:10, :10])
```

```python
# primeros 10 elementos de la diagonal principal
print("Diagonal principal de H (primeros 10 valores):")
print(np.diag(H)[:10])
```

6. <h4> Se utiliza la función numpy.linalg.eigh() para calcular los autovalores y autovectores de una matriz simétrica H. Esta función muestra:

- Psi: matriz con columnas de autovectores correspondientes a cada E_n
</h4>

```python
import numpy as np

# Cálculo de autovalores y autovectores
E, Psi = np.linalg.eigh(H)

# primeros 5 autovalores
print("Primeros 5 autovalores:")
print(E[:5])
```

1. <h4>Función de Green para la ecuación de difusión</h4>

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
D = 1
t_pulso = 5  # t'
x_prime = 0
x = np.linspace(-20, 20, 1000)
tiempos = [0, 4, 5.5, 6, 7, 8]

plt.figure(figsize=(10, 6))

# graficar G para diferentes tiempos
for t in tiempos:
    if t <= t_pulso:
        G = np.zeros_like(x) 
    else:
        coef = 1 / np.sqrt(4 * np.pi * D**2 * (t - t_pulso))
        exponent = - (x - x_prime)**2 / (4 * D**2 * (t - t_pulso))
        G = coef * np.exp(exponent)
    
    plt.plot(x, G, label=f"t = {t}")


plt.xlabel("x")
plt.ylabel("G(x, x')")
plt.title("Función de Green para la ecuación de difusión (pulso en t' = 5)")
plt.legend()
plt.grid(True)
plt.show()
```

2. <h4>Función de Green como solución de una ecuación diferencial lineal no homogénea con coeficientes variables</h4>

```python
import numpy as np
import matplotlib.pyplot as plt

# Dominio
x = np.linspace(0, 1, 500)

# Valores de xi (fuente)
xis = [0.2, 0.4, 0.6, 0.8]

plt.figure(figsize=(10, 6))

# G por tramos
for xi in xis:
    G = np.piecewise(x,
                     [x < xi, x >= xi],
                     [lambda x: x * (1 - xi),
                      lambda x: xi * (1 - x)])
    plt.plot(x, G, label=fr"$\xi = {xi}$")

# Gráfica
plt.xlabel("x")
plt.ylabel(r"$G(x, \xi)$")
plt.title(r"Función de Green de $\frac{d^2G}{dx^2} = 0$ con $G(0) = G(1) = 0$")
plt.grid(True)
plt.legend()
plt.show()
```

3. <h4>Función de Green para resolver una ecuación diferencial con condiciones frontera</h4>

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
a, b = 0, 1
n = 20
x_vals = np.linspace(0, 1, 200)
xi_vals = np.linspace(a, b, n)
f_xi = np.sin(np.pi * xi_vals)  # función fuente f(ξ)
dxi = (b - a) / (n - 1)  

# p(x) = 1, q(x) = 0
def G(x, xi):
    return np.where(x < xi, x * (1 - xi), xi * (1 - x))

# Calcular 'y' como integral numérica
y = []
for x in x_vals:
    integrando = f_xi * G(x, xi_vals)
    integral = np.sum(integrando) * dxi
    y.append(integral)

# Gráfica 
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y, label=r"$y(x)$ usando función de Green")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.title(r"Solución $y(x)$ con $f(\xi) = \sin(\pi \xi)$ usando $n=20$ puntos")
plt.grid(True)
plt.legend()
plt.show()
```

1. <h4>Función dilogaritmo (spence)</h4>

Se grafica la función evaluada sobre el plano complejo z = x + i. En el dominio x, y ∈ [-5, 5]. Se muestran dos mapas de color: uno para la parte real y otro para la parte imaginaria de spence(z). Esto permite visualizar cómo se comporta esta función compleja en el plano.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import spence

# malla de números complejos z = x + i y
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# evaluar función spence en la malla
W = spence(Z)

# parte real e imaginaria
W_real = np.real(W)
W_imag = np.imag(W)

# Gráficas
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Parte real
c1 = axs[0].pcolormesh(X, Y, W_real, shading='auto', cmap='magma')
axs[0].set_title("Parte real de spence(z)")
axs[0].set_xlabel("Re(z)")
axs[0].set_ylabel("Im(z)")
fig.colorbar(c1, ax=axs[0])

# Parte imaginaria
c2 = axs[1].pcolormesh(X, Y, W_imag, shading='auto', cmap='seismic')
axs[1].set_title("Parte imaginaria de spence(z)")
axs[1].set_xlabel("Re(z)")
axs[1].set_ylabel("Im(z)")
fig.colorbar(c2, ax=axs[1])

plt.tight_layout()
plt.show()
```

2. <h4>Función hipergeométrica y ecuaciones 18.129, 18.133 y 18.134</h4>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hyp2f1, eval_legendre, eval_chebyt, eval_chebyu

# Dominio 
x = np.linspace(-1, 1, 400)
z = (1 - x) / 2  

# Grado n 
n = 4

# Ecuación 18.129 - Legendre
P_legendre_direct = eval_legendre(n, x)
P_hyper = hyp2f1(-n, n + 1, 1, z)

# Ecuación 18.133 - Chebyshev T_n
T_direct = eval_chebyt(n, x)
T_hyper = hyp2f1(-n, n, 0.5, z)

# Ecuación 18.134 - Chebyshev U_n 
U_direct = eval_chebyu(n, x)
U_hyper = (n + 1) * hyp2f1(-n, n + 2, 1.5, z)

# Gráficas 
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

# Legendre
axs[0].plot(x, P_legendre_direct, label=r"$P_n(x)$ directo", color='tab:orange',linewidth=2)
axs[0].plot(x, P_hyper, '--', label=r"$\,_2F_1(-n,n+1;1;(1-x)/2)$", color='tab:green',linewidth=2)
axs[0].set_title("Verificación Ecuación 18.129 – Polinomio de Legendre")
axs[0].legend()
axs[0].grid(True)

# Chebyshev T_n
axs[1].plot(x, T_direct, label=r"$T_n(x)$ directo",color='firebrick',linewidth=2)
axs[1].plot(x, T_hyper, '--', label=r"$\,_2F_1(-n,n;1/2;(1-x)/2)$",color='tab:blue',linewidth=2)
axs[1].set_title("Verificación Ecuación 18.133 – Chebyshev Tₙ")
axs[1].legend()
axs[1].grid(True)

# Chebyshev U_n
axs[2].plot(x, U_direct, label=r"$U_n(x)$ directo",color='navy',alpha=0.9, linewidth=2)
axs[2].plot(x, U_hyper, '--', label=r"$(n+1)\,_2F_1(-n,n+2;3/2;(1-x)/2)$",color='tab:orange',linewidth=2)
axs[2].set_title("Verificación Ecuación 18.134 – Chebyshev Uₙ")
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

3. <h4>Función ellipj para funciones elípticas de Jacobi</h4>

Se verifica que las funciones elípticas de Jacobi cumplen las identidades clásicas 4.32, 4.33 y 4.39, comparando sus lados izquierdo y derecho mediante cálculos numéricos. La función ellipj de scipy.special permite evaluar sn(u,k), cn(u,k) y dn(u,k) de forma directa.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import ellipj

# Parámetros
k_vals = [0.3, 0.7, 0.95]   # módulos elípticos
u = np.linspace(-10, 10, 1000)

# Identidades 
for k in k_vals:
    m = k**2
    sn, cn, dn, _ = ellipj(u, m)

    # Identidad 4.32
    id1 = sn**2 + cn**2

    # Identidad 4.33
    id2 = dn**2 + k**2 * sn**2

    # errores respecto a 1
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(u, id1 - 1, label=fr"$k = {k}$", color='navy')
    plt.title("sn² + cn² - 1")
    plt.xlabel("u")
    plt.ylabel("error")
    plt.grid(True)
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(u, id2 - 1, label=fr"$k = {k}$", color='firebrick')
    plt.title("dn² + k² sn² - 1")
    plt.xlabel("u")
    plt.ylabel("error")
    plt.grid(True)
    plt.legend()

    plt.suptitle(fr"Verificación de identidades 4.32 y 4.33 para $k={k}$")
    plt.tight_layout()
    plt.show()
```

```python
# Parámetro
k = 0.7
m = k**2

# u y v
u_vals = np.linspace(-4, 4, 500)
v = 1.2  # valor fijo de v

# Evaluar funciones elípticas
sn_u, cn_u, dn_u, _ = ellipj(u_vals, m)
sn_v, cn_v, dn_v, _ = ellipj(v, m)

# lado izquierdo y derecho 
sn_sum = ellipj(u_vals + v, m)[0]
numerador = sn_u * cn_v * dn_v + sn_v * cn_u * dn_u
denominador = 1 - k**2 * sn_u**2 * sn_v**2
rhs = numerador / denominador

# Gráfica comparación
plt.figure(figsize=(8, 5))
plt.plot(u_vals, sn_sum, label=r"$\text{sn}(u+v,k)$", color='orange')
plt.plot(u_vals, rhs, '--', label="RHS (identidad 4.39)", color='tab:green')
plt.title("Verificación identidad 4.39: sn(u+v)")
plt.xlabel("u")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
plt.show()

# Error absoluto
plt.figure(figsize=(8, 4))
plt.plot(u_vals, sn_sum - rhs, label="Error", color='tab:green')
plt.title("Error: sn(u+v) - RHS")
plt.xlabel("u")
plt.ylabel("Diferencia")
plt.grid(True)
plt.legend()
plt.show()
```
